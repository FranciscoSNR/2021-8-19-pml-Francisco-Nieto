```python
import mglearn
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
# generate synthetic two-dimensional data
X, y = make_blobs(random_state=33, n_samples=1000)
# build the clustering model
#plt.scatter(X,y)
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
prediction = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=prediction, s=60)
```

```python
from sklearn.cluster import AgglomerativeClustering
agg1 = AgglomerativeClustering(n_clusters=4)
prediction1 = agg1.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=prediction1, s=60)
```

```python
agg2 = AgglomerativeClustering(n_clusters=4,linkage='average' )
prediction2 = agg2.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=prediction2, s=60)
```

```python

agg3 = AgglomerativeClustering(n_clusters=4, linkage='complete')
prediction3 = agg3.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=prediction3, s=60)
```

```python
from sklearn.cluster import DBSCAN
dbs = DBSCAN(eps=.55, min_samples=5)
prediction3 = dbs.fit_predict(X)
plt.scatter(X[:, 0], X[:, 1], c=prediction3, s=60)
```
